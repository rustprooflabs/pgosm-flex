```python

i = 0

filenames = []

for filename in os.listdir(project_folder):
    if filename.endswith('.py') and filename not in ['__init__.py']:
        input_path = os.path.join(project_folder, filename)
        outfile = os.path.join(OUTPUT_PATH, filename[:-3] + '.png')
        print(f'\nCreating flow diagram for {input_path}\n')
        print(f'\nSaving to {outfile}\n')
        code2flow.code2flow(raw_source_paths=[input_path],
                            output_file=outfile,
                            no_trimming=True,
                            hide_legend=False)

        filenames.append(input_path)
        i += 1

print(f'\nProcessed {i} files from {project_folder}\n')
#Image(outfile)

final_output_file = os.path.join(OUTPUT_PATH, 'pgosm_flex_full.png')
code2flow.code2flow(raw_source_paths=filenames,
                    output_file=final_output_file,
                    no_trimming=True,
                    hide_legend=False)
```
